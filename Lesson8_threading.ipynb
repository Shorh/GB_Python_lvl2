# Главный поток

```python
def sample(name, num):
    for idx in range(1, num+1):
        print(f'{ name } is running { idx }')
        
sample('main', 12)
```

# Потоки
Потоки перехватывают управление рандомно

```python
import threading

thread1 = threading.Thread(target=sample, args=('thread-1', 15))
thread2 = threading.Thread(target=sample, args=('thread-2', 30))
thread3 = threading.Thread(target=sample, args=('thread-3', 5))
thread4 = threading.Thread(target=sample, args=('thread-4', 20))
thread5 = threading.Thread(target=sample, args=('thread-5', 10))

thread1.start()
thread2.start()
thread3.start()
thread4.start()
thread5.start()
```

# Блокировка

```python
import threading

lock = threading.Lock()

def sample(name, num):
    lock.acquire()
    for idx in range(1, num+1):
        print(f'{ name } is locking { idx }')
    lock.release()
```

```python
thread1 = threading.Thread(target=sample, args=('thread-1', 15))
thread2 = threading.Thread(target=sample, args=('thread-2', 30))
thread3 = threading.Thread(target=sample, args=('thread-3', 5))
thread4 = threading.Thread(target=sample, args=('thread-4', 20))
thread5 = threading.Thread(target=sample, args=('thread-5', 10))

thread1.start()
thread2.start()
thread3.start()
thread4.start()
thread5.start()
```

# Конкуренция потоков

```python
import time

def get_value(name, source):
    while source:
        time.sleep(1)
        print(name, source.pop())
```

```python
import threading

source = list(range(30))

thread1 = threading.Thread(target=get_value, args=('thread-1', source))
thread2 = threading.Thread(target=get_value, args=('thread-2', source))
thread3 = threading.Thread(target=get_value, args=('thread-3', source))
thread4 = threading.Thread(target=get_value, args=('thread-4', source))
thread5 = threading.Thread(target=get_value, args=('thread-5', source))

thread1.start()
thread2.start()
thread3.start()
thread4.start()
thread5.start()
```

## Блокировка конкуренции потоков

```python
import time
import threading

lock = threading.Lock()

def get_value(name, source):
    while source:
        lock.acquire()
        if source:
            time.sleep(1)
            print(name, source.pop())
        lock.release()
```

```python
import threading

source = list(range(30))

thread1 = threading.Thread(target=get_value, args=('thread-1', source))
thread2 = threading.Thread(target=get_value, args=('thread-2', source))
thread3 = threading.Thread(target=get_value, args=('thread-3', source))
thread4 = threading.Thread(target=get_value, args=('thread-4', source))
thread5 = threading.Thread(target=get_value, args=('thread-5', source))

thread1.start()
thread2.start()
thread3.start()
thread4.start()
thread5.start()
```
